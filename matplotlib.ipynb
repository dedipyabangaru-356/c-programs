```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as mp
import numpy as np
xpoint=np.array([0,100])
ypoint=np.array([0,5])
mp.plot(xpoint,ypoint)
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np
x=np.array([0,100])

mp.plot(x,linestyle= 'dotted')
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np
xpoint=np.array([0,29,34,67,23,100])
ypoint=np.array(['me','my','self','i','why','you?'])
mp.plot(xpoint,ypoint,'o')
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np
xpoint=np.array([0,29,34,67,23,100])
ypoint=np.array(['me','my','self','i','why','you?'])
mp.bar(ypoint,xpoint)
mp.show()
```

```python
import matplotlib.pyplot as m
import numpy as n

a=n.array([70,6,4,20])
m.pie(a)
m.show()
x = n.random.normal(170, 10, 250)
m.hist(x)
m.show() 
```
